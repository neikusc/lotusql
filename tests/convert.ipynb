```python
import pandas as pd, sqlite3
df = pd.read_csv("../data/train.csv")                 # add args like sep="|", dtype=...
# df.head(4)
conn = sqlite3.connect("../data/train.sqlite")
df.to_sql("properties", conn, if_exists="replace", index=False)
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("../data/train.sqlite")
cur = conn.cursor()
# Option 2: full CREATE TABLE statement
cur.execute("SELECT sql FROM sqlite_master WHERE type='table' AND name='properties';")
create_sql = cur.fetchone()
print("\nCREATE TABLE statement:\n", create_sql[0] if create_sql else "not found")

conn.close()
```

```python
conn = sqlite3.connect("../data/train.sqlite")
cur = conn.cursor()
# Option 1: column-by-column info
cur.execute("PRAGMA table_info(properties);")
cols = cur.fetchall()
print("cid | name | type | notnull | dflt_value | pk")
for row in cols:
    print(row)
```

```python
import sqlite3

conn = sqlite3.connect("../data/train.sqlite")
cur = conn.cursor()

# Query total number of records
cur.execute("SELECT COUNT(*) FROM properties;")
count = cur.fetchone()[0]

print("Total records:", count)



cur.execute("PRAGMA table_info(properties);")
columns = [row[1] for row in cur.fetchall()]  # row[1] = column name

print(columns)
conn.close()
```
